```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.dates as mdates
```

```python
data=pd.read_csv('covid_19_india.csv')
data.tail(90)
```

```python
data.rename(columns={'State/UnionTerritory':'State_UnionTerritory'},inplace=True)
data
```

# Maharashta analysis

```python
mh=data[data.State_UnionTerritory  == 'Maharashtra']
mh
```

```python
mh.dropna(subset=["Confirmed"])
mh.head(10)
```

```python
count=mh['Confirmed']-mh['Cured']-mh['Deaths']
count
```

```python
#x_values = [datetime.datetime.strptime(d,"%m/%d/%Y").date() for d in kerala['Date']]
plt.figure(figsize= (60,30))
plt.xticks(rotation = 90 ,fontsize = 40)
plt.yticks(fontsize = 40)
            
plt.scatter(mh.Date,count,color='black')
plt.plot(mh.Date,count,color='black')
plt.scatter(mh.Date,mh.Cured, color='green')
plt.plot(mh.Date,mh.Cured, color='green')
plt.scatter(mh.Date,mh.Deaths, color='red')
plt.plot(mh.Date,mh.Deaths, color='red')
plt.title("Total corona cases in Maharashtra indicated with blue",fontsize = 40)
plt.title("Total active corona cases in Maharashtra indicated with black""----total deaths with red""----total cured with green",fontsize = 70)
plt.xlabel("Date",fontsize = 50)
plt.ylabel("No of Cases",fontsize = 50)
plt.show()
```
